```python
data1 = data.groupby('Species')['SepalLengthCm']
```

# LINE PLOT

*Ploting a line graph to showi the sepal length of the three species of Iris flowers Using different colors to represent each species*.

```python
data.duplicated().sum()
```

```python
versicolor = data1.get_group('Iris-versicolor')
virginica = data1.get_group('Iris-virginica')
setosa= data1.get_group('Iris-setosa')
```

```python
plt.figure(figsize=(8,6))

x=range(1,len(versicolor)+1)

plt.plot(x, versicolor.values, label='Iris-versicolor')
plt.plot(x, virginica.values, label='Iris-virginica')
plt.plot(x, setosa.values, label='Iris-setosa')

plt.xlabel('Index', fontsize=12)
plt.ylabel('Sepal Length (cm)', fontsize=12)
plt.title('Sepal Length of Iris Species', fontsize=14)
plt.legend()
plt.grid(True)
plt.savefig('sepal_length_plot.png')
```

# SCATTER PLOT

*Creating a scatter plot to show the relationship between the petal length and petal width using different colors to represent each species.*

```python

sns.scatterplot(data = data,x ='PetalWidthCm',y='PetalLengthCm',hue='Species', palette='magma')
plt.grid(True)
plt.title('Relation between Petalwidth & Petallengt',fontsize=16)
plt.savefig('Scatterplot of petalwidth and petalLength .png')  

plt.show()
```

# HISTOGRAM

*Ploting a histogram showing the distribution of the petal widths using different colors to represent each species and display a legend.*

```python

sns.histplot(data=data,x='PetalWidthCm',hue ='Species',bins =40,palette='magma')

plt.title('Histrogram of Petal Width')
plt.savefig('Histrogram.png')  
```

# BAR PLOT

*Creating a bar plot to show the average sepal width for each species.*

```python
ave = data.groupby('Species')['PetalWidthCm'].mean().reset_index()

sns.barplot(data=ave,x='Species',y='PetalWidthCm',hue ='Species',palette='magma')

plt.title('Average of sepalwidth for ecah species',fontsize = 16)
plt.xlabel('Species',fontsize = 13)
plt.ylabel('Average of sepalwidth',fontsize = 13)

plt.xticks(rotation = 45)
plt.savefig('Ave sepalwidth barplot.png') 

plt.show()
```

# PIE CHART

*Assuming that the dataset represents the entire population of Iris flowers to create a pie chart showing the proportion of each species in the population.*

```python
data['Species'].value_counts().plot(kind='pie',autopct="%.1f%%")

plt.title('Porportion of each species')
plt.savefig('pie plot of each species.png') 
```

# BOX PLOT

*Creating a box plot showing the distribution of sepal length for each species.*

```python
sns.boxplot(data=data, x='Species', y='PetalLengthCm',palette ='rocket')

plt.title('Box plot of each species')
plt.savefig('boplot of each species.png')
```

# SUBPLOT

*Combining the scatter plot and the histogram into one figure by using subplots.*

```python
plt.figure(figsize=(10,6))

plt.subplot(121)

sns.boxplot(data=data, x='Species', y='PetalLengthCm',palette='magma')

plt.title('Boxplot')

plt.subplot(122)

sns.scatterplot(data = data,x ='PetalWidthCm',y='PetalLengthCm',hue='Species',palette='magma')

plt.title('Scatterplot')
plt.savefig('subplot.png') 

plt.show()
```
